```python

import math
board=[' 'for  _ in range(9)]
def print_board():
  for i in range(0,9,3):
    print(f'| {board[i]} | {board[i+1]} | {board[i+2]} |')
print_board()
def is_winner(player):
  win_conditions = [
      [0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]
  ]
  for condition in win_conditions:
    if board[condition[0]] == board[condition[1]] == board[condition[2]] == player and board[condition[0]] != ' ':
      return True
  return False
def is_board_full():
  return ' ' not in board
def minmax(is_maximizing):
  if is_winner('0'):
    return 1 #player wins
  elif is_winner('x'):
    return -1 #Ai wins-that maximizes the minimum possible moves
  elif is_board_full():
    return 0 #tie

  if is_maximizing:
    best_score = -math.inf
    for i in range(9):
      if board[i] == ' ':#cwll is emptyy next simulating the move
        board[i]='0'
        score = minmax(False)
        board[i]=' '     #backtracking    making empty
        best_score = max(score,best_score)# here explorin other possible moves without affecting other states
    return best_score
  else:
    best_score = math.inf
    for i in range(9):
      if board[i] == ' ':
        board[i]='x'
        score = minmax(True)
        board[i]=' '
        best_score = min(score,best_score)
    return best_score
def best_move():
  best_score=-math.inf
  move =0

  for i in range(9):
    if board[i]==' ':
      board[i]='0'
      score = minmax(False)
      board[i]=' '
      if score > best_score:
        best_score = score
        move = i
  return move

def player_move():
  move = int(input('Enter move between (1-9):')) -1
  if board[move] == ' ':
    board[move] = 'x'
  else:
    print('Invalid move')
    player_move
print("Welcome to tic-tac-toe! you are 'x' and AI is '0'")
print_board()

while True:
  player_move()
  print_board()
  if is_winner('x'):
    print('you win')
    break
  elif is_board_full():
    print("It is a tie")
    break
  ai_move = best_move()
  board[ai_move]='0'
  print_board()

  if is_winner('0'):
    print('AI wins')
    break
  elif is_board_full():
    print("It is a tie")
    break
  move = best_move()
```
